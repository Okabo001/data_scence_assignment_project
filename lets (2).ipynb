```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score
from joblib import dump


# Loading the data from the path file in my google drive
dataset = pd.read_excel('/content/Book3.csv.xlsx')
data


#  dataframe from my dataset in spreedsheet
data = {
    'Weather': ['Clear', 'Rainy', 'Foggy', 'Snowy', 'Clear', 'Foggy', 'Rainy', 'Snowy'],
    'Road_Conditions': ['Dry', 'Wet', 'Wet', 'Icy', 'Dry', 'Wet', 'Wet', 'Icy'],
    'Location': ['Nairobi Intersection', 'Mombasa Highway', 'Kisumu Residential', 'Nakuru Roundabout', 'Eldoret Intersection', 'Nyeri Residential', 'Thika Roundabout', 'Naivasha Highway'],
    'Time_of_Day': ['Morning', 'Afternoon', 'Evening', 'Night', 'Morning', 'Morning', 'Morning', 'Morning'],
    'Accident_Severity': [3.5, 2.2, 1.7, 4, 3.2, 1.5, 2.9, 4.2],
}

dataset = pd.DataFrame(data)

# here I try Encoding the categorical variables using one-hot encoding
encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(dataset[['Weather', 'Road_Conditions', 'Location', 'Time_of_Day']])

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, dataset['Accident_Severity'], test_size=0.2, random_state=42)

#  Training the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Evaluating the model
Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared (R^2): {r2}")

# Save the trained model
dump(model, 'accident_severity_model.joblib')
```
